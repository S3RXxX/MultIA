# Using MMS for TTS task

## Imports

```python
from transformers import VitsModel, AutoTokenizer
import torch
import os
import scipy
import pygame
```

## Inference

### Loading model and tokenizer

```python
os.environ["TRANSFORMERS_VERBOSITY"] = "error"
```

```python
model = VitsModel.from_pretrained("facebook/mms-tts-eng") # spa/eng
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-eng")
```

### Doing inference

```python
text = "Hello from the other side UwU."
inputs = tokenizer(text, return_tensors="pt")

with torch.no_grad():
    output = model(**inputs).waveform
```

## Saving audio

```python
scipy.io.wavfile.write("audio_test.wav", rate=model.config.sampling_rate, data=output.float().numpy().flatten())
```

## Playing audio

### Opening audio with os

```python
os.system("audio_test.wav")
```

### Playing audio with pygame

```python
# Initialize the mixer module
pygame.mixer.init()

# Load the mp3 file
pygame.mixer.music.load("audio_test.wav")

# Play the loaded mp3 file
pygame.mixer.music.play()
```

### Playing audio with Audio from IPython.display (notebooks)

```python
from IPython.display import Audio

Audio(output.numpy(), rate=model.config.sampling_rate)
```
